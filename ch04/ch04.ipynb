```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import toeplitz
```

```python
# create some matrices
A = np.random.randn(3,4)
B = np.random.randn(100,100)
C = -toeplitz(np.arange(8),np.arange(10))

# and show them as images
fig,axs = plt.subplots(1,3,figsize=(10,3))

axs[0].imshow(A,cmap='gray')
axs[1].imshow(B,cmap='gray')
axs[2].imshow(C,cmap='gray')

for i in range(3): axs[i].axis('off')
plt.tight_layout()
# plt.savefig('Figure_04_01.png',dpi=300)
plt.show()
```

```python
"""
행렬의 곱셈에는 여러 가지가 있음. 
1. 아다마르 곱
  보통 점원()으로 표기함
2. 스칼라 곱셈
  => 1과 2는 원소별로 산술연산이 일어나게 된다! 당연히 행렬끼리의 덧셈, 뺄셈도!
3. 행렬 곱셈(표준 행렬 곱셈)
  - 아다마르 곱이나 스칼라 곱과 구분하기 위해. "표준"이라는 단어를 붙였음
  - 그냥 AB 이렇게 표기
"""

A = np.array([[1, 2], [3, 4]])
B = np.array([[1, 2], [3, 4]])
scalar = 3

# 행렬의 원소별 곱
print(scalar * A)
# 행렬의 아다마르곱
A * B
np.multiply(A, B)
```

```python
# Exercise 2
# 원본 행렬
A = np.arange(100).reshape(10, 10)
# 부분 행렬
B = A[:5, :5]

_, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(A, cmap='gray', origin='upper', vmin=0, vmax=np.max(A))
axs[0].plot([4.5,4.5],[-.5,9.5],'w--')
axs[0].plot([-.5,9.5],[4.5,4.5],'w--')
axs[0].set_title('Original matrix')
# text labels
for (j, i), num in np.ndenumerate(A):
  axs[0].text(i, j, num, color=[.8, .8, .8], ha='center', va='center')

axs[1].imshow(B,cmap='gray',origin='upper',vmin=0,vmax=np.max(A))
axs[1].set_title('Submatrix')
# text labels
for (j,i),num in np.ndenumerate(B):
  axs[1].text(i,j,num,color=[.8,.8,.8],ha='center',va='center')


plt.show()
```
